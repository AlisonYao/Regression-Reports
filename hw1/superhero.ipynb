```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/zhuang/Desktop/movieChinaData3.csv')
df.head()
```

```python
budgets = df['Budget (Mil$)']
grosses = df['Chinese Gross (Mil$)']
```

```python
profitRatio = grosses/budgets
```

```python
plt.scatter(df["Year"], profitRatio)
plt.show()
```

```python
for i in range(len(profitRatio)):
    if profitRatio[i] >= 1:
        print(df['Movie Title'][i])
```

```python
color = []
for i in range(len(profitRatio)):
    if profitRatio[i] >= 1:
        color.append('red')
    else:
        color.append('blue')
```

```python
plt.scatter(df["Year"], profitRatio, c=color)
plt.xticks([2002, 2004, 2006, 2008, 2010, 2012, 2014, 2016, 2018, 2020])
plt.axhline(y=1.0, color='r', linestyle=':')
plt.xlabel("Year")
plt.ylabel('ratio')
plt.show()
```
